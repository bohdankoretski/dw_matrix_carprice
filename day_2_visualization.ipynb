```
def group_plot_data(feat_groupby, feat_agg='price_value', agg_funcs=[np.mean, np.median, np.size], feat_sort='mean', top=50, subplots=True):
  return (
    df
 .groupby(feat_groupby)[feat_agg]
 .agg(agg_funcs)
 .sort_values(by=feat_sort, ascending=False)
 .head(top)

).plot(kind='bar', figsize=(22, 8), subplots=subplots)
```

```
group_plot_data('param_marka-pojazdu');
```

```
group_plot_data('param_kraj-pochodzenia', feat_sort='size');
```

```
group_plot_data('param_kolor', feat_sort='mean');
```
